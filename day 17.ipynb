```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv(r'C:\Users\pavan kumar\Documents\pandasdata\mtcars.csv')
```

```python
data
```

```python
df=data['hp']
```

```python
df
```

```python
mean=np.mean(df)
std=np.std(df)
print(mean,std)
```

```python
threshold = 1.9
outlier = [] 
for i in df: 
    z = (i-mean)/std 
    if z > threshold: 
        outlier.append(i) 
outlier
```
